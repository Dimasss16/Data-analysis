```python
# Create separate dataframes for hits and non-hits.
hit_songs_df = bs[bs['popularity'] == 'hit']
mediocre_songs_df = bs[bs['popularity'] == 'mediocre']
hit_songs_df
```

```python
mediocre_songs_df
```

## weeks on board

```python
# Calculating central tendency of weeks on board for hits
print("Mean: ", hit_songs_df['weeks-on-board'].mean())
print("Median: ", hit_songs_df['weeks-on-board'].median())
print("Mode: ", hit_songs_df['weeks-on-board'].mode())
```

```python
hit_songs_df['weeks-on-board'].describe()
```

```python
# Calculating central tendency of weeks on board for non-hits
print("Mean: ", mediocre_songs_df['weeks-on-board'].mean())
print("Median: ", mediocre_songs_df['weeks-on-board'].median())
print("Mode: ", mediocre_songs_df['weeks-on-board'].mode())
```

```python
def variance(xs):
    mu = (sum(xs))/len(xs)
    n = len(xs)
    n = n-1 if n in range(1, 30) else n  
    square_deviation = lambda x : (x - mu) ** 2 
    return sum( map(square_deviation, xs) ) / n
```

```python
xs = list(mediocre_songs_df['weeks-on-board'])
variance(xs)
```

## Interpretation

The central tendency values for the number of weeks that hits remain on the chart are more than double those of regular songs. Therefore, we can conclude that hits, on average, stay on the chart twice as long. However, we also observe that the median for hits is less than the average, indicating that a greater number of them spend less time on the chart (positive asymmetry coefficient).

Interestingly, the mode for hits is closer to both the median and the mean, which is not the case for regular songs. If the mode value significantly deviates from other central tendency measures, it suggests a high dispersion of numerical data (variance $> 30$).

## peak-rank

```python
# Calculating central tendency of peak-rank for hits
print("Mean: ", hit_songs_df['peak-rank'].mean())
print("Median: ", hit_songs_df['peak-rank'].median())
print("Mode: ", hit_songs_df['peak-rank'].mode())
```

```python
# Calculating central tendency of peak-rank for regular songs
print("Mean: ", mediocre_songs_df['peak-rank'].mean())
print("Median: ", mediocre_songs_df['peak-rank'].median())
print("Mode: ", mediocre_songs_df['peak-rank'].mode())
```

## Interpretation

Nearly all the hit songs reached first place on the chart at some point. This is indicated by the alignment of the median and mode, with a minor deviation from the arithmetic mean. However, the data on common songs shows a considerable variation, evident in the significant difference between the mode and the mean.

# Table description

We create a new table, artist_hit_df, based on the existing data. It should contain the number of hits for each artist's works.

```python
artist_hit_df = pd.DataFrame({'artist': list_of_artists, 'number of hits': num_of_hits})
artist_hit_df.sort_values(by = 'number of hits', ascending = False) 
```

```python
print("Mean: ", artist_hit_df['number of hits'].mean())
print("Median: ", artist_hit_df['number of hits'].median())
print("Mode: ", artist_hit_df['number of hits'].mode())
```

## Interpretation

As mentioned before, hits are only a small portion of all the songs in the chart. Therefore, on average, each artist has three or more hits. However, the median is 0, indicating that over half of the artists have no hits, a sign of positive asymmetry. Despite this, we observe maximum values of 300 and 363. Based on this data, we can conclude that the median equals the mode when more than half of the values are the same.

We created a boxplot to show the number of weeks that 'hit' songs spent on the chart.

The interquartile range, calculated by subtracting the 1st quartile from the 3rd quartile, is 11. Therefore, values that lie more than $11 \cdot 1.5 = 16.5$ away from the 1st or 3rd quartile are considered outliers. The upper bound is 22, so any value above 38.5 will be marked as outliers (represented as black dots on the plot). The lower bound is 11, indicating that at least 75% of hits remained on the chart for 11 weeks or more.

# Choosing feature columns

We only select numerical columns. From these, we choose pairs that can be interpreted well. This is based on the idea that a hit song will likely secure a high position on the chart, both at the time of release and over time. Furthermore, if a song is currently high on the chart, it probably debuted at the top or immediately upon release, rather than starting from the bottom.

```python

numeric_col = ['rank','last-week','peak-rank','weeks-on-board'] 


corr_matrix = bs.loc[:,numeric_col].corr()
print(corr_matrix)


sn.heatmap(corr_matrix, annot=True)
```

# Conclusions

To sum up:

- Rank correlates with last-week, as songs that initially rank low on the chart are likely to remain there, while top tracks tend to retain their positions.
- Rank also correlates with peak-rank. The data indicates that songs reaching high chart positions at some point tend not to lose these positions significantly.
- Peak-rank correlates with last-week. This suggests that if a song originally ranks high, it's more likely to climb higher.
- Weeks-on-board is negatively correlated with peak-rank, implying popular songs which reach high positions don't stay on the chart for long.
- A weak negative correlation exists between weeks-on-board and last-week, indicating that songs reaching high chart positions early on don't stay there for long.
