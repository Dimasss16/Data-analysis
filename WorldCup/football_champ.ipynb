# Data description

The <a href="https://github.com/rfordatascience/tidytuesday/blob/master/data/2022/2022-11-29/worldcups.csv">table</a> shows data about the footbal world champuonships from 1930 year on.



Columns:

   **host** -- Host country (str)
   
   **winner** -- Winner country (str)
   
   **second** --  Country whose team secured the second place (str)
   
   **third** -- Country whose team secured the third place (str)
   
   **fourth** -- Country whose team secured the fourth place (str)
   
   **goals_scored** -- Total number of goals in the championship (int)
   
   **teams** -- Number of participants in the championship (int)
   
   **games** -- Total number of games in the championship (int)
   
   **attendance** -- Number of spectators (int)

# Attendance in 20th and 21st century

## Proposition

- Football games attendance will be lower in the 20th century than in the 21st century

```python
df[df['year'] < 2000]['attendance'].mean() # в 20м веке
```

```python
df[df['year'] >= 2000]['attendance'].mean() # в 21м веке
```

## Interpretation

In the 21st century, average attendance at the World Cup has almost doubled compared to the 20th century. This is mainly due to progress in organization, increased interest in soccer, and the presence of more top-tier players.

# Average number of goals scored

## Proposition

- As footbal evolves, the average number of goals at championships will increase

```python
df1 = df
df1
```

```python
df1['20s'] = df1['year']//20
df1
```

```python
avg_goals = df1.groupby('20s')['goals_scored'].mean()
avg_goals
```

```python
sns.lineplot(data = avg_goals)
plt.show()
```

## Interpretation

The line graph shows that the number of goals scored in championships has nearly doubled over time.

The hypothesis was partially confirmed. It can be observed that the number of goals decreased during a certain period of time (from 1954 to 1966). However, it then continued to grow and more than doubled the initial number.

# Correlation analysis

## Proposition

- As the number of games increases, the attendance will also climb.

```python
df1['games'].corr(df1['attendance'])
```

## Interpretation

The correlation coefficient is close to one, which indicates a relationship between the variables where an increase in one indicator leads to an increase in the other. In other words, the more games there are at a championship, the higher the attendance.

# Average attendance for each host country

## Proposition

- The Brazilian championship will have the highest attendance, as Brazil was one of the early favorites.

- The championship organized by Uruguay will have the lowest attendance, as it was the first World Cup.

```python
avg_attendance = df1.groupby('host')['attendance'].mean()
avg_attendance
```

```python
avg_at = pd.DataFrame(avg_attendance)
avg_at
```

```python
avg_at = avg_at.reset_index()
avg_at
```

```python
fig = plt.figure()

sns.barplot(x='host', y ='attendance', ci=None, data = avg_at)
plt.xticks(rotation=45)

fig.set_figwidth(12)
fig.set_figheight(7)

plt.show()
```

## Interpretation

As a result, we see the highest attendance in developed countries such as the USA, Russia, Japan, and Germany, and the lowest attendance at the Uruguayan championship. The hypothesis is partially confirmed.

# The number of hosted championships

## Proposition

- Most countries would organize only one World Cup.
- Brazil and Germany are the only countries that have organized multiple World Cups.

```python
num_of_tournaments = df1.groupby(['host'])['host'].count()
num_of_tournaments
```

```python
num_of_tournaments_df = pd.DataFrame(num_of_tournaments)
num_of_tournaments_df
```

```python
num_of_tournaments_df.rename(columns = {'host' : 'number of hosted champs'}, inplace = True)
num_of_tournaments_df
```

```python
num_of_tournaments_df = num_of_tournaments_df.reset_index()
num_of_tournaments_df
```

```python
num_of_tournaments_df.sort_values(by = 'number of hosted champs', ascending = False)
```

## Interpretation

Based on the data in the table, it can be seen that Brazil, France, Germany, Italy, and Mexico have all hosted the World Cup more than once. However, France, Germany, and Italy have only won one of their World Cups, and no country has managed to become champions in both of their World Cups.

# Won championships

## Proposition

- The most number of times a world champion would be Brazil.
- Uruguay would win the least number of times.

```python
winners = list(set(list(df1['winner'])))
winners
```

```python
df1
```

```python
num_of_wins = pd.DataFrame(df1.groupby(['winner'])['winner'].count())
num_of_wins
```

```python
num_of_wins.rename(columns = {'winner' : 'number of wins'}, inplace = True)
num_of_wins 
```

```python
num_of_wins = num_of_wins.reset_index()
num_of_wins
```

```python
num_of_wins.sort_values(by = 'number of wins', ascending = False)
```

```python
data = list(num_of_wins['number of wins'])
keys = list(num_of_wins['winner'])

palette_color = sns.color_palette("Blues")
plt.pie(data, labels=keys, colors=palette_color, autopct='%.0f%%')
  

plt.show()
```

## Interpretation

The pie chart shows that Brazil has won the most championships. The hypothesis is partially confirmed.
